```python
# Task 2 Univariate analysis

#Q. what does the distribution of customer ratings look likes? is it skewed?

sns.distplot(df['Rating'])

# to draw vertical line for mean

plt.axvline(x= np.mean(df['Rating']), c = 'red', ls = '--', label = 'mean')
plt.axvline(x=np.percentile(df['Rating'],25),c= 'green', ls = '--', label = '25-75th percentile' )
plt.axvline(x=np.percentile(df['Rating'],75),c= 'green', ls = '--' )

# user rating is uniform
```

```python
# Task 4 dealing with duplicate rows and missing values

df.duplicated()

# found 3 True values
```

```python
# to get the number of duplicated rows

df.duplicated().sum()

# output = 3
```

```python
df[df.duplicated() == True]   # it will return the duplicated rows if there any
# 3 rows
```

```python
# remove duplicate rows

# inplace = True is used for permanent change

df.drop_duplicates(inplace=True)   

# 3 rows removed
```

```python
# check for missing values

df.isna().sum()


# member = 73 , Sports and Travel = 39, ..
```

```python
sns.heatmap(df.isnull(), cbar = False)

```

```python
# fill the missing values


# df.fillna(0) # we can fill missing values with zero
df.fillna(df.mean(), inplace = True)  # it will replace only numerical values with mean
```

```python
# so, using mode to fill missing values

# df.mode()

df.mode().iloc[0]
```

```python
# filled the missing value using following code
# again run heat map to check, no missing values now
df.fillna(df.mode().iloc[0], inplace = True)  
```
